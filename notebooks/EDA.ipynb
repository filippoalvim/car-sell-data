# Análise Exploratória de Dados — Anúncios de Venda de Carros

```python
import pandas as pd
import plotly.express as px

car_data = pd.read_csv('../vehicles.csv')
car_data.head()
```

```python
car_data.info()
```

```python
car_data.describe()
```

## Histograma — Distribuição da Quilometragem

```python
fig = px.histogram(car_data, x='odometer', nbins=50,
                   title='Distribuição da Quilometragem',
                   labels={'odometer': 'Quilometragem (milhas)'})
fig.show()
```

## Gráfico de Dispersão — Preço vs Quilometragem

```python
fig = px.scatter(car_data, x='odometer', y='price',
                 title='Preço vs Quilometragem',
                 labels={'odometer': 'Quilometragem (milhas)', 'price': 'Preço (USD)'},
                 opacity=0.5)
fig.show()
```

## Histograma — Distribuição de Preços por Condição

```python
fig = px.histogram(car_data, x='price', color='condition', nbins=50,
                   title='Distribuição de Preços por Condição',
                   labels={'price': 'Preço (USD)', 'condition': 'Condição'})
fig.show()
```
